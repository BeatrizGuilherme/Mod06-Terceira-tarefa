#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc.loc[sinasc['APGAR5']<=3,'APGAR5_cat']='asfixia grave'
sinasc.loc[(sinasc['APGAR5']>3)&(sinasc['APGAR5']<=5),'APGAR5_cat']='asfixia moderada'
sinasc.loc[(sinasc['APGAR5']>5)&(sinasc['APGAR5']<=7),'APGAR5_cat']='asfixia leve'
sinasc.loc[sinasc['APGAR5']>7,'APGAR5_cat']='normal' 
```

```python
sinasc.head()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
ax= sns.boxplot(data=sinasc
                ,y='APGAR5'
                ,x ='GESTACAO'
                , orient= 'v'
                , hue= 'SEXO'
               )
ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
plt.show() 
```

```python
plt.close('all')
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabela_cruzada= pd.crosstab(sinasc['GESTACAO'],sinasc['APGAR5'])
tabela_cruzada
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
print(tabela_cruzada.sum(1))
print()
tabela_cruzada_perc = tabela_cruzada.div(tabela_cruzada.sum(axis=1), axis = 0)

print(tabela_cruzada_perc)
```
